```python
class Persona:

    def __init__(self, nombre, nacionalidad):
        self.nombre = nombre
        self.nacionalidad = nacionalidad
class Estudiante(Persona):
    pass
            
class Trabajador(Persona):
    pass

#Creamos objetos de las clases HIJAS
alumno = Estudiante("Sultan", "Paraguaya")
empleado = Trabajador("Conrad", "Paraguaya")

#Mostramos los atributos heredados
print(f"El estudiante se llama {alumno.nombre} y su nacionalidad es {alumno.nacionalidad}.")
print(f"El trabajador se llama {empleado.nombre} y su nacionalidad es {empleado.nacionalidad}.")
```

```python
class Heroe:

    def __init__(self, nombre,habilidad):
        self.nombre = nombre
        self.habilidad = habilidad
class Villano(Heroe):
    pass

aprendiz = Heroe('Martin', 'velocidad y fuerza')
experto = Villano('Jonathan', 'volar y controlar mentes')

print(f'yo me llamo {aprendiz.nombre} y tengo la habilidad de {aprendiz.habilidad}')
print(f'yo me llamo {experto.nombre} y yo tengo la habilidad de {experto.habilidad}')
```
